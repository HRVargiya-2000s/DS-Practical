```python
%pip install fitter --quiet
from fitter import Fitter
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print("\n1 - Load Data")
data = [170, 165, 180, 172, 160, 175, 168, 155, 185, 190, 162, 178, 168, 172, 180, 160, 165, 172, 168, 175]
data = np.array(data)
print("Loaded Data:", data)

print("\n2 - Fit Distributions")
f = Fitter(data, distributions=['norm', 'expon', 'lognorm', 'gamma', 'beta'])
f.fit()
print("Distributions fitted.")

print("\n3 - Show Parameter Summary")
print(f.summary())

print("\n4 - Best-Fit Distribution")
print(f.get_best())

print("\n5 - Plot Histogram and PDF")
f.plot_pdf()
plt.show()
```
