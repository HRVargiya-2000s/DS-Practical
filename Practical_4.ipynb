```python
import numpy as np
import matplotlib.pyplot as plt

print("\n1 - Normal Distribution")
normal_dist = np.random.normal(0, 1, 1000)
print(np.mean(normal_dist))
print(np.std(normal_dist))

print("\n2 - Plot Normal Distribution PDF")
plt.hist(normal_dist, bins=30, density=True, alpha=0.5)
x = np.linspace(-4, 4, 100)
plt.plot(x, 1/(np.std(normal_dist)*np.sqrt(2*np.pi))*np.exp(-(x-np.mean(normal_dist))**2/(2*np.std(normal_dist)**2)), linewidth=2)
plt.show()

print("\n3 - Plot Normal Distribution CDF")
plt.hist(normal_dist, bins=30, density=True, alpha=0.5, cumulative=True)
plt.show()

print("\n4 - Poisson Distribution")
poisson_dist = np.random.poisson(5, 1000)
print(np.mean(poisson_dist))
print(np.var(poisson_dist))

print("\n5 - Plot Poisson Distribution PDF")
plt.hist(poisson_dist, bins=15, density=True, alpha=0.5)
plt.show()

print("\n6 - Plot Poisson Distribution CDF")
plt.hist(poisson_dist, bins=15, density=True, alpha=0.5, cumulative=True)
plt.show()
```
