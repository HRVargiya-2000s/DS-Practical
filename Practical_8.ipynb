```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print("\n1 - Load Data")
data = {
    'Age': [15, 16, 15, 17, 16, 16, 15, 16],
    'StudyHours': [5, 4, 6, 3, 4, 7, 5, 6],
    'PreviousGrade': [60, 70, 55, 80, 65, 90, 62, 75],
    'Result': [0, 1, 0, 1, 1, 1, 0, 1]
}
df = pd.DataFrame(data)
print(df)

print("\n2 - Split Data: Features/Target")
X = df[['Age', 'StudyHours', 'PreviousGrade']]
y = df['Result']
print("Features:\n", X)
print("Target:\n", y)

print("\n3 - Split into Train/Test Sets")
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
print("Train Features:\n", X_train)
print("Test Features:\n", X_test)

print("\n4 - Train Decision Tree Model")
model = DecisionTreeClassifier(random_state=0)
model.fit(X_train, y_train)
print("Model trained.")

print("\n5 - Predict and Evaluate")
y_pred = model.predict(X_test)
print("Predicted:", y_pred)
print("Actual   :", list(y_test))
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

print("\n6 - Plot Decision Tree")
plt.figure(figsize=(10,6))
plot_tree(model, feature_names=['Age', 'StudyHours', 'PreviousGrade'], class_names=['Average', 'Clever'], filled=True)
plt.title("Decision Tree for Student Classification")
plt.show()
```
