```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print("\n Step 1: Load Dataset ")
df = pd.read_excel("Practical-1.xlsx")
print(df.head())


print("\n Step 2: Dataset Info ")
print(df.info())

print("\n Step 3: Dataset Summary ")
print(df.describe(include="all"))

print("\n Step 4: Scatter Plot (Age vs Income) ")
sns.scatterplot(data=df, x="Age", y="Income", hue="Gender")
plt.title("Scatter Plot: Age vs Income")
plt.show()

print("\n Step 5: Box Plot (Income by Gender) ")
sns.boxplot(data=df, x="Gender", y="Income")
plt.title("Income Distribution by Gender")
plt.show()

print("\n Step 6: Box Plot (Income by Education Level) ")
sns.boxplot(data=df, x="Education Level", y="Income")
plt.xticks(rotation=45)
plt.title("Income Distribution by Education Level")
plt.show()

print("\n Step 7: Contingency Table (Education vs Marital Status) ")
contingency = pd.crosstab(df["Education Level"], df["Marital Status"])
print(contingency)

print("\n Step 8: Histogram (Age Distribution by Education Level) ")
sns.histplot(data=df, x="Age", hue="Education Level", multiple="stack")
plt.title("Age Distribution by Education Level")
plt.show()

print("\n Step 9: Stacked Bar Chart (Employment Status by Gender) ")
employment_gender = pd.crosstab(df["Gender"], df["Employment Status"])
print(employment_gender)
employment_gender.plot(kind="bar", stacked=True)
plt.title("Employment Status by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```
