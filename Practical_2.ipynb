```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print("1 Load Dataset")
df = pd.read_excel("Practical-2.xlsx")
print(df.head())


print("\n2 - Measures of Central Tendency")
print("Test1 - Mean:", df['Test1 Score'].mean(), "Median:", df['Test1 Score'].median(), "Mode:", df['Test1 Score'].mode().values)
print("Test2 - Mean:", df['Test2 Score'].mean(), "Median:", df['Test2 Score'].median(), "Mode:", df['Test2 Score'].mode().values)

print("\n3 - Correlation")
correlation = df['Test1 Score'].corr(df['Test2 Score'])
print("Correlation Coefficient:", correlation)

print("\n4 - Scatter Plot")
plt.scatter(df['Test1 Score'], df['Test2 Score'])
plt.xlabel('Test1 Score')
plt.ylabel('Test2 Score')
plt.title('Scatter plot: Test1 vs Test2')
plt.show()

print("\n5 - Percentiles, Deciles, Quartiles")
for col in ['Test1 Score', 'Test2 Score']:
    print(f"\n{col}:")
    print("25th percentile:", np.percentile(df[col], 25))
    print("50th percentile (Median):", np.percentile(df[col], 50))
    print("75th percentile:", np.percentile(df[col], 75))
    print("30th decile:", np.percentile(df[col], 30))
    print("40th decile:", np.percentile(df[col], 40))
    print("70th decile:", np.percentile(df[col], 70))
    print("1st quartile:", np.percentile(df[col], 25))
    print("3rd quartile:", np.percentile(df[col], 75))

print("\n6 - Box Plot")
plt.boxplot([df['Test1 Score'], df['Test2 Score']], labels=['Test1', 'Test2'])
plt.title('Box plot for Test1 and Test2')
plt.ylabel('Scores')
plt.show()

print("\n7 - Measures of Variation")
for col in ['Test1 Score', 'Test2 Score']:
    print(f"\n{col}:")
    print("Range:", df[col].max() - df[col].min())
    print("IQR:", np.percentile(df[col], 75) - np.percentile(df[col], 25))
    print("Variance:", df[col].var())
    print("Standard Deviation:", df[col].std())

print("\n8 - Measures of Shape (Skewness and Kurtosis)")
for col in ['Test1 Score', 'Test2 Score']:
    print(f"\n{col}:")
    print("Skewness:", df[col].skew())
    print("Kurtosis:", df[col].kurtosis())
```
