```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print("\n1 - Load Data")
data_dict = {
    'Area': [2000, 1800, 2200, 1500, 2400, 1900, 2100],
    'Bedrooms': [3, 4, 3, 1, 4, 3, 4],
    'Bathrooms': [2, 2, 2, 1, 3, 2, 3],
    'GarageCars': [2, 1, 2, 1, 3, 2, 3],
    'SalePrice': [250000, 230000, 265000, 200000, 330000, 275000, 295000],
}
df = pd.DataFrame(data_dict)
print(df)

print("\n2 - Split Data: Features/Target")
X = df[['Area', 'Bedrooms', 'Bathrooms', 'GarageCars']]
y = df['SalePrice']
print("Features:\n", X)
print("Target:\n", y)

print("\n3 - Split into Train/Test Sets")
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print("Train Features:\n", X_train)
print("Test Features:\n", X_test)

print("\n4 - Train Linear Regression Model")
model = LinearRegression()
model.fit(X_train, y_train)
print("Model trained.")

print("\n5 - Predict and Evaluate")
y_pred = model.predict(X_test)
print("Predicted values:", y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error:", mse)
print("R^2 Score:", r2)

print("\n6 - Coefficients and Intercept")
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

print("\n7 - Plot Predicted vs Actual")
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Sale Price")
plt.ylabel("Predicted Sale Price")
plt.title("Actual vs Predicted Sale Price")
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red')
plt.show()
```
