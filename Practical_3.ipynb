```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("1 - NumPy Array Creation")
arr_5x5 = np.random.randint(1, 101, (5,5))
print(arr_5x5)

print("\n2 - NumPy Statistics on 1x25 Array")
arr_1x25 = arr_5x5.reshape(1,25)
print("Mean:", np.mean(arr_1x25))
print("Median:", np.median(arr_1x25))
print("Variance:", np.var(arr_1x25))
print("Standard Deviation:", np.std(arr_1x25))

print("\n3 - Percentile/Decile/Quartile (NumPy)")
randarr = np.random.randint(1,101,10)
print("Array:", randarr)
print("25th percentile:", np.percentile(randarr, 25))
print("50th percentile:", np.percentile(randarr, 50))
print("75th percentile:", np.percentile(randarr, 75))
print("10th decile:", np.percentile(randarr, 10))
print("Quartiles:", np.percentile(randarr, [25,50,75]))

print("\n4 - Matplotlib Bar Chart")
x = ['A', 'B', 'C', 'D']
y = [10, 20, 30, 40]
plt.bar(x, y, color='skyblue')
plt.title('Simple Bar Chart')
plt.xlabel('Labels')
plt.ylabel('Values')
plt.show()

print("\n5 - Matplotlib Pie Chart")
labels = ['Red', 'Blue', 'Green', 'Yellow']
sizes = [20, 30, 10, 40]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['red','blue','green','yellow'])
plt.title('Pie Chart Example')
plt.show()

print("\n6 - Pandas DataFrame Load")
df = pd.read_csv('sales_data.csv')
print(df.head())

print("\n7 - Pandas Statistics (Units Sold, Revenue)")
print("Units Sold - Mean:", df['quantity'].mean(),"Median:",df['quantity'].median(),"Variance:",df['quantity'].var(),"Std:",df['quantity'].std())
print("Revenue - Mean:", df['total_sales'].mean(),"Median:",df['total_sales'].median(),"Variance:",df['total_sales'].var(),"Std:",df['total_sales'].std())

print("\n8 - Pandas Group Statistics by Product")
grouped = df.groupby('product')[['quantity','total_sales']].agg(['mean','median','var','std'])
print(grouped)

print("\n9 - Units Sold vs Revenue Scatterplot")
plt.scatter(df['quantity'], df['total_sales'], alpha=0.6)
plt.xlabel('Units Sold')
plt.ylabel('Revenue')
plt.title('Units Sold vs Revenue')
plt.show()
```
